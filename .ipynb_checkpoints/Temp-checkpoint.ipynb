By looking at the distribution of `acousticness` for our purposes Zellner's is not suitable; normal prior would not be optimal here because of the non-symmetric distribution, whereas for `loudness` and `valence` the normal distribution seems to be fairly good estimate. 

Next we will select suitable candidate prior distributions for the variables. We need to give `acousticness` special treatment in order to be able to select a satisfying custom informative prior:

### Acousticness

```python
x = np.linspace(sp.stats.gamma.ppf(0.01, 2), sp.stats.gamma.ppf(0.999, 2), 100)
loc = -.5

fig = plt.figure()
fig.set_figwidth(10)
plt.xlim(-4,5)
sns.set_context('paper',rc={"lines.linewidth": 2.5})
sns.distplot(global_data[predictors[0]], label='Prior distribution', )
for k in np.linspace(1.2,4, num=5):
    plt.plot(X-1, [sp.stats.gamma.pdf(x,k, loc=loc) for x in X], '--', lw=1,
             label='Ga(a={0:.3})'.format(k))
    
fig.legend()
```

As we see from here, Gamma distribution with parameter value $a = 1.9$ seems to be able to include most of the information. We relax the scaling a bit in order to allow more variation and choose $a = 2.5$ as the parameter value. We will also include the location parameter $b = -0.5$ into the model.  

### Loudness

```python
X = np.linspace(-8,8,num=100)

fig = plt.figure()
fig.set_figwidth(10)
sns.distplot(global_data[predictors[1]], label='Prior distribution')
for  in np.linspace(0.7,3, num=4):
    plt.plot(X, [sp.stats.norm.pdf(x, scale=) for x in X], '--', 
             lw=1, label='N(0, {0:.3})'.format(**2))
    
fig.legend()
```

We can see that even with small $\sigma^2$ values, the normal distribution covers underlying kde quite well. However there seems to be a bit more mass on the left tail of the kde, and thus we will relax variance further. Taking this fact into consideration it seems like  $N(\mu = 0, \sigma^2 = 5)$ is a good candidate for informative prior that at the same time has wide enough scale variable to avoid strong bias and is still able to capture the essence of the underlying distribution.

### Valence

```python
X = np.linspace(-8,8,num=100)

fig = plt.figure()
fig.set_figwidth(10)
sns.distplot(global_data[predictors[2]], label='Prior distribution')
sns.set_style()
for  in np.arange(1,5):
    plt.plot(X, [sp.stats.norm.pdf(x, scale=) for x in X],  '--', 
             lw=1, label='N(0,{0})'.format(**2))
    
fig.legend()
```

Here the  $N(\mu = 0, \sigma^2 = 4)$ seems like a good candidate for the prior. Akin to the previous feature, here too the justifications are very similar - we can put it in another way; we want an informative prior that can find just the right balance between overfitting (too little variation) and underfitting (too much variation). Model should be able to generalize into the real population distribution.
